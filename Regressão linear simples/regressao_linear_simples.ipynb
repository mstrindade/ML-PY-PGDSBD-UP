```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston # para carregar os dados
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # importa o modelo
```

```python
x = np.linspace(0, 10, 100)
y = (5*x+2 + 20*np.random.rand(1, len(x)))
x2 = np.linspace(20, 25, 50)
y2 = (5*x2+2 + 20*np.random.rand(1, len(x2)))
x = np.append(x,x2)
y = np.append(y,y2)
```

```python
plt.scatter(x, y, color = 'red')
```

```python
X = pd.DataFrame(x)
X[1] = 1
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
regr = LinearRegression() # cria o modelo
regr.fit(X_train, y_train) # treina o modelo
```

```python
r2_train = regr.score(X_train, y_train)
r2_test = regr.score(X_test, y_test)
print('R2 no set de treino: %.2f' % r2_train)
print('R2 no set de teste: %.2f' % r2_test)
```

```python
b = regr.intercept_
```

```python
a = regr.coef_[0]
```

```python
xp = np.linspace(0, 25, 100)
yp = a*xp + b

PointX = 15
PointY = regr.predict([[PointX, 1]])

plt.scatter(x, y, color = 'red')
plt.plot(xp, yp, color = 'blue')
plt.scatter(PointX, PointY, color = 'blue', s=500)
plt.show()
```
